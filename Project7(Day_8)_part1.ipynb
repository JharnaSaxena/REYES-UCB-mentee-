First, let's define some basic functions:

```python
pi= np.pi

def sqrt(x):
  return np.sqrt(x) # square root This func returns the squareroot of x
def exp(x):
  return np.exp(x) # exponential This func returns the exponential of x
def sin(x):
  return np.sin(x) # sine curve This func returns the sine curve w.r.t. x
def cos(x):
  return np.cos(x) # cosine curve This func returns the cosine curve w.r.t. x

def log(x):
  return np.log(x) # logarithm  This func returns the log value w.r.t. x

def poly0(x):
  return (pow(x,3)/6.) + (pow(x,4)/8.0) # polynomial (combination of powers)
```

# Exercise #1: Defining the derivative

```python
def fin_deriv(f, x, a):
	"""
	this defines the rate of change of the function f as a function of x

	the final point is at xf = x + a
	the initial point is at xi = x - a
	the difference between these two is xf - xi = 2*a
	"""

	# your code here
	return (f(x+a)-f(x-a))/(2*a)
```

You can play around with this to understand the impact of changing "a" -- finer "resolution" gets more accurate results.

```python
x = np.arange(0.3, 12, 0.00001)

a = 0.5 # try values between 0.1 and 2

# define sine function
def sin(x):
  f = np.sin(x)
  return f

# take the derivative
dsin = fin_deriv(sin, x, a)

plt.plot(x, np.sin(x), label="sin")
plt.plot(x, dsin, color='black', label = "numerical derivative")
plt.plot(x, np.cos(x), '--', color='red', label="actual derivative")
plt.xlabel('x')
plt.legend()
plt.grid(True)
plt.show()
```

# Exercise #2: Test convergence

Now let's make some plots as a function of the resolution "a" and see what our error function looks like.

```python
# define our error function
def error_fun(Xvec,Yvec):

  """ This function takes a dependent variable (in our case, the derivative) and independent variable (in our case, a) and returns the error function. """

  num = Yvec[0:-1]-Yvec[1:]
  dnum = Xvec[0:-1]-Xvec[1:]
  Ytmp = 100 * abs(num/dnum)

  Xtmp = Xvec[0:-1]

  return Xtmp, Ytmp

def master_plot_vs_a(ylabel1,ydata1,ylabel2,ydata2,xtmp,xlabel0,ydata3):

  """ This will format our plots for us, so we don't have to type this out every time. """

  plt.figure(figsize=(9.0,5))
  plt.subplot(211)
  plt.xticks(fontsize=15)
  plt.yticks(fontsize=15)

  plt.ylabel(ylabel1,size=25)

  plt.axhline(y=ydata2,color='k',linewidth=1)
  plt.plot(a, ydata1, color='r')

  plt.axvline(x=0,color='k',linewidth=1)

  plt.subplot(212)
  plt.xticks(fontsize=15)
  plt.yticks(fontsize=15)

  plt.axhline(y=0,color='k',linewidth=1)
  plt.axvline(x=0,color='k',linewidth=1)

  plt.ylabel(ylabel2,size=25)
  plt.plot(xtmp, ydata3, color='g')

  plt.xlabel(xlabel0,size=25, position=(1,1.2))
  plt.grid(True)
  plt.show()


x = 1
a=np.arange(pow(10,-5),.1,pow(10,-3))


"""
******************************
******************************
    exponential
******************************
******************************
"""

# your code here

# take the derivative of the exponential function

# calculate the error function

# plot the numerical derivative

# plot the error function
deri=fin_deriv(np.exp,x,a)
sig=error_fun(a,deri)
xtmp,ydata3=error_fun(a,deri)
ylabel1="Graph!"
ydata2=np.exp(1)
ydata1=deri
ylabel2="Error here!"
xlabel0=""

master_plot_vs_a(ylabel1,ydata1,ylabel2,ydata2,xtmp,xlabel0,ydata3)


"""
******************************
******************************
    sin
******************************
******************************
"""

# your code here

# take the derivative of the exponential function

# calculate the error function

# plot the numerical derivative

# plot the error function
deri=fin_deriv(np.sin,x,a)
sig=error_fun(a,deri)
xtmp,ydata3=error_fun(a,deri)
ylabel1="Graph!"
ydata2=np.sin(1)
ydata1=deri
ylabel2="Error here!"
xlabel0=""

master_plot_vs_a(ylabel1,ydata1,ylabel2,ydata2,xtmp,xlabel0,ydata3)


"""
******************************
******************************
    cos
******************************
******************************
"""
# your code here

# take the derivative of the exponential function

# calculate the error function

# plot the numerical derivative

# plot the error function
deri=fin_deriv(np.cos,x,a)
sig=error_fun(a,deri)
xtmp,ydata3=error_fun(a,deri)
ylabel1="Graph!"
ydata2=np.cos(1)
ydata1=deri
ylabel2="Error here!"
xlabel0=""

master_plot_vs_a(ylabel1,ydata1,ylabel2,ydata2,xtmp,xlabel0,ydata3)


"""
******************************
******************************
    poly
******************************
******************************
"""

# your code here

# take the derivative of the exponential function

# calculate the error function

# plot the numerical derivative

# plot the error function
deri=fin_deriv(poly0,x,a)
sig=error_fun(a,deri)
xtmp,ydata3=error_fun(a,deri)
ylabel1="Graph!"
ydata2=poly0(1)
ydata1=deri
ylabel2="Error here!"
xlabel0=""

master_plot_vs_a(ylabel1,ydata1,ylabel2,ydata2,xtmp,xlabel0,ydata3)

"""
******************************
******************************
    log
******************************
******************************
"""

# your code here

# take the derivative of the exponential function

# calculate the error function

# plot the numerical derivative

# plot the error function
deri=fin_deriv(log,x,a)
sig=error_fun(a,deri)
xtmp,ydata3=error_fun(a,deri)
ylabel1="Graph!"
ydata2=log(1)
ydata1=deri
ylabel2="Error here!"
xlabel0=""

master_plot_vs_a(ylabel1,ydata1,ylabel2,ydata2,xtmp,xlabel0,ydata3)

```
