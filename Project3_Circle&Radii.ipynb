```python
import numpy as np
import matplotlib.pyplot as plt
n=10#no of points we wanna have more the points more the chances of having better outcomes and statistical ibservations
x=np.random.uniform(-1,1,n)#random generation from -1 to 1 and n number of points
y=np.random.uniform(-1,1,n)#random generation from -1 to 1 and n number of points
d=np.sqrt(x**2 + y**2)#calculate distance between x and y point
inside=d<1#this checks for less than 1 radi
listy = inside.tolist()#transform array into list and then check using for loop i didnt used while loop as i tend to induce infinite loops
cinside = 0#count of d<1
for i in listy:
    if i == True:
        cinside += 1#if d<1 found it increment the value by 1
pi = 4 * cinside / n#used to calc pi i didnt knew this i had to search and get formula
print("Estimated value of pi : ",pi)
plt.scatter(x[inside], y[inside], color="blue", s=15, label="inner")#for inncer circle
plt.scatter(x[~inside], y[~inside], color='red', s=15, label='outer')#for outer circle
circle = plt.Circle((0, 0), 1, color='black', linestyle='--')#outline
plt.title(f"Estimated pi ={pi} ")
plt.xlabel("x")
plt.ylabel("y")
#lables
plt.legend()
plt.grid(True)#grid cause they look good
plt.show()#atlast showing the chart
#NOW, lets do same for n=10000 or more to see why it matters to have more number of samples
n1=10000#no of points we wanna have more the points more the chances of having better outcomes and statistical ibservations
x=np.random.uniform(-1,1,n1)#random generation from -1 to 1 and n number of points
y=np.random.uniform(-1,1,n1)#random generation from -1 to 1 and n number of points
d=np.sqrt(x**2 + y**2)#calculate distance between x and y point
inside=d<1#this checks for less than 1 radi
listy = inside.tolist()#transform array into list and then check using for loop i didnt used while loop as i tend to induce infinite loops
cinside = 0#count of d<1
for i in listy:
    if i == True:
        cinside += 1#if d<1 found it increment the value by 1
pi = 4 * cinside / n#used to calc pi i didnt knew this i had to search and get formula
print("Estimated value of pi : ",pi)
plt.scatter(x[inside], y[inside], color="blue", s=15, label="inner")#for inncer circle
plt.scatter(x[~inside], y[~inside], color='red', s=15, label='outer')#for outer circle
circle = plt.Circle((0, 0), 1, color='black', linestyle='--')#outline
plt.title(f"Estimated pi ={pi} ")
plt.xlabel("x")
plt.ylabel("y")
#lables
plt.legend()
plt.grid(True)#grid cause they look good
plt.show()#atlast showing the chart
```
