Position formula x(t) = x + vt + 0.5a*t^2

```python
def position(x,v,t,a):#calc position after time t and inital pos x intial velocity v and accelartion as its value given by the user a
  return x + (v*t )+ (0.5)*a*(t**2)#formula position formula wrt x axis
def velocity(x,t,a):#calculates the final velocity after time t and inital velocity x and accelartion a
  return x+a*t
def Motion_vs_time(x,y,vx,vy,tm=10,dt=0.1):
#+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
#here arr represnt the arry of time periods or timestamps
#+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
    arr=np.arange(0,tm,dt)#make arr named array with time values from 0 to tm and 0.1 setps that meant dt that mean change or differentialtion thing with respect to time
    x=position(arr,x,vx,0)#calc horizontal pos wrt time with no accelartion
    y=position(arr,y,vy,-9.8)#acc=9.8 also vertical postion and vertical accelartion due to gravity
    vx=velocity(arr,vx,0)#horizontal veltocity remains contant as the uh accelartion is zero
    vy=velocity(arr,vy,-9.8)#calc of vertical velo over time due to gravity
    # Position x vs time
    plt.subplot(221)#we use subplot to divide it into part one etc it seems like a quardant except with just all graphs and yeah we can also write plt.subplot(2,2,1) but in short form we wrote plt.subplot(221)
    plt.plot(arr,x)#this usually plots the line chart as i read that in my high school python classes by default its a line chart
    plt.xlabel("time")#labels of x
    plt.ylabel("x position")#label of y
    plt.grid(True)#grids because they look cool to my data science oriented brain
    plt.title("horizontal Position vs time")#name of the chart
    # Position y vs time
    plt.subplot(222)#subplotted ont the second part
    plt.plot(arr,y)#the plotting of line chart of arr and y
    plt.xlabel("time")#labels of x
    plt.ylabel("y position")#label of y
    plt.grid(True)#same
    plt.title("vertical Position vs time")#name
    # Velocity x vs time
    plt.subplot(223)#subplot
    plt.plot(arr, vx)
    plt.xlabel("time")
    plt.ylabel("vx")
    plt.grid(True)
    plt.title("horizontal Velocity vs time")
    # Velocity y vs time
    plt.subplot(224)
    plt.plot(arr, vy)
    plt.xlabel("time")
    plt.ylabel("vy")
    plt.grid(True)
    plt.title('Vertical Velocity vs Time')
    plt.show()

# Example :
Motion_vs_time(x=40,y=20,vx=80,vy=100,tm=20)

def plot_trajectory(x, y, vx, vy, tm, dt=0.1):
    t = np.arange(0, tm, dt)#this  create array with name t with time values staring from 0 to the tm time and also with dt steps that is 0.1 because its the differentation thing and that works with change in the position etc
    x = position(t, x, vx, 0)#calc position of the variable
    y = position(t, y, vy, -9.8)#clac position of y variable
    if tm>=0:#checks if value of y is above ground level or not because pojecticle cant be made above ground
      plt.figure(figsize=(8, 6))#crate new figure window i didnt studied that but watched this too
      plt.plot(x, y)#same plotting line chart
      plt.xlabel('Horizontal Position x (m)')
      plt.ylabel('Vertical Position y (m)')
      plt.title('Projectile Trajectory')
      plt.grid(True)
      plt.show()
    else:
      print("Uh oh")#if the uh projectile value is below ground level it will return "uh oh" because it  cant perform plotting

# Example usage:
plot_trajectory(x=40,y=20,vx=80,vy=100,tm=20)
```
