The projectile motion game of ball and monster


```python
def trajectory(v,angel,x=0,y=0,g=9.81):
  newnew=np.radians(angel)#conversion of  degree to radian
  vx=v*np.cos(newnew)#calc horizontal component
  vy=v*np.sin(newnew)#calc vertical component
  t=np.arange(0,10,0.1)#create array of timestamps with 0.1 difference
  x=x+vx*t
  y=y+vy*t-(0.5*g*(t**2))
  #theabove calc x and y position over time
  return x,y #this return bydefault as tuple which are immuatable and x,y represent projectile path
def optimum(v,ex,ey,x=0,y=0,g=9.8):#func that calc optimum efficeient angle to hit target monster
  dx=ex-x
  dy=ey-y
  #calc difference from start to target monster
  a=(g*(dx**2))/(2*(v**2))
  b=-(dx)
  c=a-dy
  d=b**2-4*a*c#discriminant
  #quadratic eq
  if d<0:
    return None
  angel1=(-b + np.sqrt(d)) / (2 * a)
  angel2=(-b - np.sqrt(d)) / (2 * a)
  #the above 2 lines of code calc the  roots of quadratic
  an=np.degrees(np.arctan(angel1))
  gel=np.degrees(np.arctan(angel2))
  #gives tangent vale of angles in degree
  if min(an,gel)>0:
    return min(an,gel)
  else:
    return max(an,gel)
    #choose minimum of both angle if found if not return largest
def plot(v,angel,target=None):#to plot trajectory or projectile
  x,y=trajectory(v,angel)#uses the trajectory func to get value of x,y ready and then plot them
  plt.plot(x,y)#by default a line chart
  plt.scatter(0,0,color="yellow",label="Shoot!")#a scatter plot
  if target:
    plt.scatter(target[0],target[1],color="green",s=60,label="target")#if target is shot properly gives green dot and else nothing
    plt.title("Projectile ball vs monster")
    plt.grid(True)
    plt.xlabel("Horizontal distance")
    plt.ylabel("vertical distance")
    plt.show()
if __name__ == "__main__":
    v= 30 #inital velocity
    ex, ey = 20,5#target pos
    angel = optimum(v, ex, ey)
    if angel is not None:#if valid angel is found
        print(f"Optimal angle to hit target at ({ex}, {ey}): {angel:.2f}Â°")
        plot(v,angel,target=(ex, ey))
        plot(v,120,target=(ex, ey))#angle is fixed to 30 degree
#i am sorry but #3 extra challeges seemed hard to me so i am just doing these
```
